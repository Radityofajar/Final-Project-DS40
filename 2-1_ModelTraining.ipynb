## Modeling

### 1. Model Selection

```python
# Import the necessary regression models from sklearn
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor,GradientBoostingRegressor, AdaBoostRegressor
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from sklearn.model_selection import KFold, GridSearchCV, cross_val_score, cross_validate
from sklearn.metrics import make_scorer, mean_squared_error, mean_absolute_error, r2_score

# Run this If you use Intel CPU to boost Sklearn model training and inference speed
# from sklearnex import patch_sklearn
# patch_sklearn()

import numpy as np
import csv
import os
from datetime import timedelta
import time
```

### 3. K Fold Cross Validation with Grid Search Hyper Parameter Tuning

```python
# Set random_seed, which will be used for both random state and seeding
random_seed = 42

# Define the models with random state set to random_seed
models = {
    'RandomForest': RandomForestRegressor(random_state=random_seed),
    'XGBoost': XGBRegressor(random_state=random_seed),
    'LinearRegression': LinearRegression(),
    'GradientBoostingRegressor': GradientBoostingRegressor(random_state=random_seed),
    'AdaBoostRegressor': AdaBoostRegressor(random_state=random_seed),
    'LGBMRegressor': LGBMRegressor(random_state=random_seed)
}

# Define hyperparameters for Grid Search
param_grids = {
    'RandomForest': {
        'n_estimators': [50, 100, 200],
        'max_depth': [10, 20, 30],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'max_features': ['auto', 'sqrt', 'log2']
    },
    'XGBoost': {
        'n_estimators': [50, 100, 200, 500],
        'max_depth': [3, 5, 7],
        'learning_rate': [0.001, 0.01, 0.1],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0]
    },
    'LinearRegression': {},
    'GradientBoostingRegressor': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.001, 0.01, 0.1],
        'max_depth': [3, 5, 7],
        'min_samples_split': [2, 5, 10],
        'min_samples_leaf': [1, 2, 4],
        'subsample': [0.6, 0.8, 1.0],
        'max_features': ['auto', 'sqrt', 'log2']
    },
    'AdaBoostRegressor': {
        'n_estimators': [50, 100, 200, 500],
        'learning_rate': [0.001, 0.01, 0.1, 1.0],
        'loss': ['linear', 'square', 'exponential']
    },
    'LGBMRegressor': {
        'n_estimators': [100, 200, 500],
        'learning_rate': [0.001, 0.01, 0.1],
        'max_depth': [3, 5, 7, -1],
        'num_leaves': [31, 50, 100],
        'min_child_samples': [10, 20, 30],
        'subsample': [0.6, 0.8, 1.0],
        'colsample_bytree': [0.6, 0.8, 1.0]
    }
}
```

```python
# Save on google collab

# from google.colab import drive
# drive.mount('/content/drive')

def record_model(model_name, results, best_model):
  # Create a directory in Google Drive
  directory = "Model/"
  file = f'{directory}record_result.csv'
  # check if it is exist and mkdir if not
  if not os.path.exists(directory):
      os.makedirs(directory)
  with open(file, 'a', newline='\n') as csvfile:
      writer = csv.writer(csvfile)
      if csvfile.tell() == 0:
          writer.writerow(["Model", "Duration", "Best Params", "MSE", "RMSE", "MAE", "R2"])
      writer.writerow([model_name, results])
  # Save best model with pickle in same directory
  pkl.dump(best_model, open(f'{directory}model_name.pkl', 'wb'))
```

```python
def test_model(model, model_name, param_grid, X_train, y_train):
    """
    Perform grid search with cross-validation and multiple regression metrics (MSE, RMSE, MAE, R2).

    Args:
    - model: The machine learning model to test.
    - param_grid: The hyperparameter grid for GridSearchCV.
    - X_train: Training feature set.
    - y_train: Training target labels.

    Returns:
    - A dictionary of mean scores for MSE, RMSE, MAE, R2.
    """

    cv = KFold(n_splits=5, shuffle=True, random_state=42)

    scoring = {
        'MSE': 'neg_mean_squared_error',
        'MAE': 'neg_mean_absolute_error',
        'R2': 'r2'
    }
    start = time.perf_counter()
    if param_grid:
        grid_search = GridSearchCV(model, param_grid, scoring=scoring, refit='R2', cv=cv, n_jobs=-1)
        grid_search.fit(X_train, y_train)
        best_model = grid_search.best_estimator_
        best_params = grid_search.best_params_
        #save model with the best param
    else:
        best_model = model
        best_model.fit(X_train, y_train)
        best_params = "Default parameters"
    duration = timedelta(seconds=time.perf_counter()-start)

    # Perform cross-validation with the best model using all scoring metrics
    cv_results = cross_validate(best_model, X_train, y_train, scoring=scoring, cv=cv, n_jobs=-1, return_train_score=False)

    mean_mse = -np.mean(cv_results['test_MSE'])
    mean_rmse = np.sqrt(mean_mse)  # RMSE
    mean_mae = -np.mean(cv_results['test_MAE'])
    mean_r2 = np.mean(cv_results['test_R2'])

    scores = {
        'Duration': duration,
        'Best Params': best_params,
        'MSE': mean_mse,
        'RMSE': mean_rmse,
        'MAE': mean_mae,
        'R2': mean_r2,
    }

    record_model(model_name=model_name, results=scores, best_model=best_model)
    return scores, best_model
```

```python
for model_name, model in models.items():
    param_grid = param_grids.get(model_name, {})
    results = test_model(model, model_name, param_grid, X_train, y_train)
    # Print the results
    print(f"Cross-Validation Results: {model_name}")
    for metric, score in results.items():
        if metric=='Best Params' or metric=='Duration':
            print(f"{metric}: {score}")
        else:
            print(f"{metric}: {score:.4f}")
```
